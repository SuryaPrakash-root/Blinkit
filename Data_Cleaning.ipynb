```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Customer_feedback = pd.read_excel('Blinkit.xlsx', sheet_name='blinkit_customer_feedback', engine='openpyxl')
Customers = pd.read_excel('Blinkit.xlsx', sheet_name='blinkit_customers', engine='openpyxl')
Market_performance = pd.read_excel('Blinkit.xlsx', sheet_name='blinkit_marketing_performance', engine='openpyxl')
Order_Items = pd.read_excel('Blinkit.xlsx', sheet_name='blinkit_order_items', engine='openpyxl')
Orders = pd.read_excel('Blinkit.xlsx', sheet_name='blinkit_orders', engine='openpyxl')
Products = pd.read_excel('Blinkit.xlsx', sheet_name='blinkit_products', engine='openpyxl')
```

```python
# Display the first few rows of each dataframe to verify successful loading
print("Customer Feedback:")
print(Customer_feedback.head())
print("Customers:")
print(Customers.head())
print("Market Performance:")
print(Market_performance.head())
print("Order Items:")
print(Order_Items.head())
print("Orders:")
print(Orders.head())
print("Products:")
print(Products.head())
```

```python
Market_performance.isnull().sum()
Market_performance.describe()
```

```python
# Finding missing values in each dataframe
dataframes = {
    "Customer_feedback": Customer_feedback,
    "Customers": Customers,
    "Market_performance": Market_performance,
    "Order_Items": Order_Items,
    "Orders": Orders,
    "Products": Products
}
dataframe = {id: df.isnull().sum() for id, df in dataframes.items()}
for id, missing in dataframe.items():
    print(f'\n Missing values in {id}:')
    print(missing)
```

```python
# Push Cleaned Dataframe to SQL Database
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine
load_dotenv()
db_host = os.getenv('DB_HOST')
db_user = os.getenv('DB_USERNAME')
db_password = os.getenv('DB_PASSWORD')
db_name = os.getenv('DB_NAME')
db_port = os.getenv('DB_PORT')
engine = create_engine(f'postgresql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}')
Customer_feedback.to_sql('Customer_Feedback', engine, if_exists='replace', index=False)
Customers.to_sql('Customers', engine, if_exists='replace', index=False)
Market_performance.to_sql('Market_Performance', engine, if_exists='replace', index=False)
Order_Items.to_sql('Order_Items', engine, if_exists='replace', index=False)
Orders.to_sql('Order', engine, if_exists='replace', index=False)
Products.to_sql('Products', engine, if_exists='replace', index=False)
engine.dispose()


```
